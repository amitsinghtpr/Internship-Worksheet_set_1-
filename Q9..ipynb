```python
#Importing important library.
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
#Importing Chrome-webriver
driver=webdriver.Chrome(r'"C:\Users\amits\Desktop\chromedriver_win32\chromedriver.exe"')
```

```python
#Maximazing windows
driver.maximize_window()
```

```python
#Opening Youtube Website
url='https://www.youtube.com/watch?v=qySOXAJW12g&t=7s'
driver.get(url)
```

**Scrapping Comments**

```python
Comments=[]
comment=driver.find_elements(By.ID,"content-text")
for i in comment[:500]:
    Comments.append(i.text)
    
Comments
```

```python
len(Comments)
```

```python
Upvote=[]
vote=driver.find_elements(By.XPATH,"//span[@class='style-scope ytd-comment-action-buttons-renderer']")
for i in vote[0:500]:
    Upvote.append(i.text)
    
Upvote
```

```python
len(Upvote)
```

```python
Time=[]
time=driver.find_elements(By.XPATH,"//a[@class='yt-simple-endpoint style-scope yt-formatted-string']")
for i in time[19:519]:
    Time.append(i.text)
    
Time
```

```python
len(Time)
```
