Write a python program to access the search bar and search button on images.google.com and scrape 10
images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.

```python
#Importing important library.
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import time
from bs4 import BeautifulSoup
```

```python
#Importing Chrome-webriver
driver=webdriver.Chrome(r'"C:\Users\amits\Desktop\chromedriver_win32\chromedriver.exe"')
```

```python
driver.maximize_window()
```

```python
link='https://www.google.co.in/imghp?hl=en&ogbl'
driver.get(link)
```

```python
search_box=driver.find_element(By.XPATH,"//input[@class='gLFyf']").send_keys("Fruits")
```

```python
srch_btn=driver.find_element(By.XPATH,"//span[@class='z1asCe MZy1Rb']").click()
```

**Scrapping Fruits Image**

```python
Fruits=[]
fruits=driver.find_elements(By.XPATH,"//img[@class='rg_i Q4LuWd']")
for i in fruits[0:11]:    
    Fruits.append(i.get_attribute('src'))
    
```
