```python
#Importing important library.
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import time
from bs4 import BeautifulSoup
import requests
```

```python
#Importing Chrome-webriver
driver=webdriver.Chrome(r'"C:\Users\amits\Desktop\chromedriver_win32\chromedriver.exe"')
```

```python
#Maximazing windows
driver.maximize_window()
```

```python
#Opening Amazon Website
url='https://www.forbes.com/billionaires/'
driver.get(url)
```

```python
Rank=[]
rank=driver.find_elements(By.XPATH,"//div[@class='rank']")
for i in rank[0:40]:
    Rank.append(i.text)
    
Rank
```

```python
Name=[]
name=driver.find_elements(By.XPATH,"//div[@class='personName']")
for i in name[0:40]:
    Name.append(i.text)
    
Name
```

```python
Net_worth=[]
net=driver.find_elements(By.XPATH,"//div[@class='netWorth']")
for i in net[0:40]:
    Net_worth.append(i.text)
    
Net_worth
```

```python
Country=[]
country=driver.find_elements(By.XPATH,"//div[@class='countryOfCitizenship']")
for i in country[0:40]:
    Country.append(i.text)
    
Country
```

```python
Age=[]
age=driver.find_elements(By.XPATH,"//div[@class='age']")
for i in age[0:40]:
    Age.append(i.text)
    
Age
```

```python
Source=[]
source=driver.find_elements(By.XPATH,"//span[@class='source-text']")
for i in source[0:40]:
    Source.append(i.text)
    
Source
```

```python
Industry=[]
ind=driver.find_elements(By.XPATH,"//div[@class='category']")
for i in ind[0:40]:
    Industry.append(i.text)
    
Industry
```

```python
df=pd.DataFrame({"Rank":Rank,"Name":Name,"Age":Age,"Country of Origin":Country,"Networth":Net_worth,"Source":Source,"Industry":Industry})
df
```

```python
df.to_csv(r'C:\Users\amits\Downloads\Forbes 40.csv',header=True,index=False )
```
